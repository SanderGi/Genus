### Define Method to Find Kuratowski Minors of a Graph



```python
K5 = graphs.CompleteGraph(5)
K33 = graphs.CompleteBipartiteGraph(3, 3)
```

```python
def findKuratowskiMinorsIfAny(G):
    result = {'K5': None, 'K33': None}
#     planar, kuratowski_subgraph = G.is_planar(kuratowski=True)
#     if planar:
#         return result
#     try:
#         result['K5'] = kuratowski_subgraph.minor(K5)
#         result['K33'] = G.minor(K33)
#     except:
#         result['K33'] = kuratowski_subgraph.minor(K33)
#         result['K5'] = G.minor(K5)
    try: 
        result['K5'] = G.minor(K5)
    except:
        pass
    try: 
        result['K33'] = G.minor(K33)
    except:
        pass
    return result
```

### Run and Plot on a Sample Graph

### Run on all Myrvold's Minors and Save to 'KuratowskiMinors.csv'



```python
with open('KuratowskiMinors.csv', 'w') as f:
    f.write("graph_number,k5_minor,k33_minor\n")
    for i, G in enumerate(minor_obs):
        minors = findKuratowskiMinorsIfAny(G)
        f.write(f'''{i},"{minors['K5']}","{minors['K33']}"\n''')
        f.flush()
```
